Now, let's create a 2D array that populates "true" and "false" values with that function.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def settest(r, i):
    c = complex(r,i)
    z = 0
    for x in xrange(0,50):
        z = z*z + c
        if (z.real > 2 or z.imag > 2):
            return False
    return True


r = 1000
xset = []
yset = []
for x in xrange(0,r):
    for y in xrange(0,r):
        xf = float(x)/r*3-1.5
        yf = float(y)/r*3-1.5
        if (settest(xf, yf)):
            xset.append(xf)
            yset.append(yf)
plt.scatter(xset, yset)
plt.title("Mandelbrot Set")
plt.show()
```
